Function that takes dataset as input and return accuracy 

```python
def calaculate_accuracy(x):
    X=pd.get_dummies(x)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.33, random_state=30)
    lgt_R = LogisticRegression(random_state = 101, solver='lbfgs', multi_class='auto')
    lgt_R.fit(X_train, y_train)
    print('Accuracy:',lgt_R.score(X_test,y_test))
    choromosomes_acc.append(lgt_R.score(X_test,y_test))
```

create a list of all the columns name except the last column 'ReadmissionWithin_90Days' , which is assighned as label

```python
columnsName=df.drop(labels= 'ReadmissionWithin_90Days', axis= 1).columns.values.tolist()
```

create 10 different chromosomes each having 30 different columns name

```python
chromosomes=[]
for i in range(10):
    chromosomes.append(random.sample(columnsName, 30))
```

function to perform mutation that takes a list of columns are randomly select a column from dataframe and replace it with a column in a list

```python
def mutation(lst):
    df_col=random.sample(columnsName, 1)
    chrom_col=random.sample(lst, 1)
    indx=lst.index(chrom_col[0])
    lst[indx]=df_col[0]
    return lst
```

function to compare the current accuracies with the previous accuracies, so that to check out if the accuracy coming is constant or not

```python
first_ten_best_accuracies=[]
def compare_acc():
    v=np.diff(first_ten_best_accuracies)
    if(all(i >=0.01 for i in v)):
        return False
    else:
        return True
```
